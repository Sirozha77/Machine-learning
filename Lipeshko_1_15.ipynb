```python
#1 Створіть одновимірний масив (вектор) з найбільшими простими числами, не перевищуючи 30, та виведіть його значення.
def is_prime(num):
    """Функція для перевірки, чи є число простим."""
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def generate_primes_up_to_n(n):
    """Генерує прості числа, не більше заданого n."""
    primes = [i for i in range(2, n+1) if is_prime(i)]
    return primes

max_value = 30


prime_numbers = generate_primes_up_to_n(max_value)


print("Масив найбільших простих чисел, не більше", max_value, ":", prime_numbers)
```

```python
#2 Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями квадратів найближчого більшого парного числа для кожного натурального числа, починаючи з 1, та виведіть його значення.
import math

def next_even_square(num):
    """Знаходить квадрат найближчого більшого парного числа."""
    even_num = math.ceil(math.sqrt(num))
    if even_num % 2 == 1:
        even_num += 1
    return even_num**2


rows, cols = 3, 3


matrix = [[0] * cols for _ in range(rows)]


current_num = 1
for i in range(rows):
    for j in range(cols):
        matrix[i][j] = next_even_square(current_num)
        current_num += 1


for row in matrix:
    print(row)
```

```python
# 3 Створіть масив розміром 6x6, заповніть його випадковими цілими числами від -50 до 50 та знайдіть добуток непарних чисел.
import numpy as np

# Розмір матриці
rows, cols = 6, 6

# Ініціалізація матриці випадковими цілими числами від -50 до 50
matrix = np.random.randint(-50, 51, size=(rows, cols))

# Виведення матриці
print("Матриця:")
print(matrix)

# Знаходження добутку непарних чисел
product_of_odd_numbers = 1
for row in matrix:
    for num in row:
        if num % 2 != 0:
            product_of_odd_numbers *= num

# Виведення добутку непарних чисел
print("\nДобуток непарних чисел:", product_of_odd_numbers)
```

```python
# 11 Створіть матрицю розміром 2x3 та вектор розміром 3, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть суму елементів матриці, які менше за відповідний елемент вектору.
import numpy as np

# Розмір матриці та вектора
rows, cols = 2, 3

# Ініціалізація матриці випадковими дійсними числами від 0 до 1
matrix = np.random.rand(rows, cols)

# Ініціалізація вектора випадковими дійсними числами від 0 до 1
vector = np.random.rand(cols)

# Виведення матриці та вектора
print("Матриця:")
print(matrix)

print("\nВектор:")
print(vector)

# Знаходження суми елементів матриці, які менше за відповідний елемент вектору
sum_below_vector = np.sum(matrix[matrix < vector.reshape(1, -1)])

# Виведення результату
print("\nСума елементів матриці, які менше за відповідний елемент вектору:", sum_below_vector)
```

```python
# 14 Створіть матрицю розміром 7x7, заповніть її випадковими дійсними числами в діапазоні від -10 до 10 та знайдіть добуток елементів матриці, які більше за їх середнє арифметичне значення.
import numpy as np

# Розмір матриці
rows, cols = 7, 7

# Ініціалізація матриці випадковими дійсними числами від -10 до 10
matrix = np.random.uniform(-10, 10, size=(rows, cols))

# Виведення матриці
print("Матриця:")
print(matrix)

# Знаходження середнього арифметичного значення
average_value = np.mean(matrix)

# Знаходження добутку елементів матриці, які більше за середнє арифметичне
product_of_greater_than_average = np.prod(matrix[matrix > average_value])

# Виведення результату
print("\nДобуток елементів матриці, які більше за середнє арифметичне:")
print(product_of_greater_than_average)
```

```python
# 15 Створіть матрицю розміром 9x9, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить найбільше за модулем ціле значення в кожному рядку матриці.
import numpy as np

# Розмір матриці
rows, cols = 9, 9

# Ініціалізація матриці випадковими дійсними числами від -0.5 до 0.5
matrix = np.random.uniform(-0.5, 0.5, size=(rows, cols))

# Виведення матриці
print("Матриця:")
print(matrix)

# Знаходження вектора-стовпчика з найбільшими за модулем цілими значеннями в кожному рядку
max_abs_values_vector = np.max(np.abs(matrix), axis=1)

# Виведення результату
print("\nВектор-стовпчик з найбільшими за модулем цілими значеннями в кожному рядку:")
print(max_abs_values_vector)
```
